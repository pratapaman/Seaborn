# Jointplot 
jointplot create a multi-panel figure that shows both the bivariate(or joint) relationship between two variables along with the Univariate (or marginal) distribution of each on separate axes.

we can use jointplot in the following 4 manners.

1. ScatterPlots
2. Hexbin Plots
3. KdePlots
4. RegPlots

```python
tips= sns.load_dataset('tips')
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip',hue='sex', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip',hue='sex', kind='kde', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='hex', data=tips)
```

```python
sns.jointplot(x='total_bill', y='tip', kind='reg', data=tips)
```

# Pairplots
